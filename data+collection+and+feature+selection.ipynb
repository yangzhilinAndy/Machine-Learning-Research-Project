```python
def get_ndays_sample(total_df,name_list,new_df,n):
    x=np.array(name_list)
    it = np.nditer(x,flags=['multi_index','refs_ok'])
    groups=total_df.groupby('serial_number')
    while not it.finished:
        ix=it.multi_index
        obj=groups.get_group(x[ix])
        if len(obj)<n:
            it.iternext()
            continue
        obj=obj.iloc[len(obj)-n:len(obj)]
        new_df=new_df.append(obj)
        it.iternext()
    return new_df
```

```python
def get_healthy_train(total_df,name_list,train_df,n):
    x=np.array(name_list)
    it = np.nditer(x,flags=['multi_index','refs_ok'])
    groups=total_df.groupby('serial_number')
    while not it.finished:
        ix=it.multi_index
        obj=groups.get_group(x[ix])
        train_size=int(len(obj)*0.7)
        if train_size<n:
            it.iternext()
            continue
        obj=obj.iloc[0:train_size+1]
        samples=obj.sample(12)
        train_df=train_df.append(samples)
        it.iternext()
    return train_df
```

```python
def get_healthy_test(total_df,name_list,train_df):
    x=np.array(name_list)
    it = np.nditer(x,flags=['multi_index','refs_ok'])
    groups=total_df.groupby('serial_number')
    while not it.finished:
        ix=it.multi_index
        obj=groups.get_group(x[ix])
        test_size=int(len(obj)*0.3)
        samples=obj.iloc[-1:-1-test_size:-1]
        samples=samples.sample(3)
        train_df=train_df.append(samples)
        it.iternext()
    return train_df
```

```python
def feature_extraction(dataset,features):
    return dataset[features]
```

```python
def change_rate(features,n,dataset,newdf):
    for feature in features:
        a=dataset[feature][1:]
        b=dataset[feature][:-1]
        c=a-b
        newdf=pd.concat([newdf,c],axis=1)
    dataset=pd.concat([dataset,newdf],axis=1)
    return dataset
```
